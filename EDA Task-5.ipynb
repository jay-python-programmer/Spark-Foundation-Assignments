### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.info()
```

### Missing Values

```python
missingno.bar(df,figsize=(16, 10))
```

```python
df.describe()
```

```python
sns.countplot(x='Ship Mode',data=df)
```

```python
sns.countplot(x='Region',data=df)
```

```python
df['Sub-Category'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(16,4))
plt.subplot(131)
sns.distplot(df['Sales'])
plt.subplot(132)
sns.distplot(df['Quantity'])
plt.subplot(133)
sns.distplot(df['Profit'])
plt.show()
```

```python
plt.figure(figsize=(16,4))
df['State'].value_counts().head(20).plot(kind='bar')
```

```python
df.groupby('Quantity')[['Segment','Quantity']].sum().plot(kind='bar')
```

```python
df.groupby('Ship Mode')[['Ship Mode','Profit']].sum().plot(kind='bar')
```

```python
df.groupby('Ship Mode')[['Ship Mode','Discount']].mean().plot(kind='bar')
```

```python
df.groupby(by=['Sub-Category'])['Sales'].sum().plot.bar(df.groupby(by=['Sub-Category'])['Sales'].sum(),figsize=(20,8))
plt.ylabel("Sales")
```

```python
df.groupby(by=['Sub-Category'])['Profit'].sum().plot.bar(df.groupby(by=['Sub-Category'])['Profit'].sum(),figsize=(20,6))
```

```python
df.groupby(by=['Category'])['Profit'].sum().plot.bar(df.groupby(by=['Category'])['Profit'].sum(),figsize=(16,6))
```

```python
df.groupby(by=['Category'])['Sales'].sum().plot.bar(df.groupby(by=['Category'])['Sales'].sum(),figsize=(16,6))
```

```python
sns.heatmap(df.corr(),fmt='.4g',annot=True)
```
